```python
# Aplicaţi algoritmul K-means clustering pentru următoarele puncte, cu K=3:
# P1: 45 85
# P2: 50 43
# P3: 40 80
# P4: 55 42
# P5: 200 43
# P6: 48 40
# P7: 195 41
# P8: 43 87
# P9: 190 40
# Listaţi cei 3 centroizi şi punctele asociate fiecărui cluster

import random
import numpy as np
p1=[45,85]
p2=[50,43]
p3=[40,80]
p4=[55,42]
p5=[200,43]
p6=[48,40]
p7=[195,41]
p8=[43,87]
p9=[190,40]
point_array=[]
d_array=[]
point_array.extend([p1,p2,p3,p4,p5,p6,p7,p8,p9])
centroizi_aux=[]
cluster=[[],[],[]]
centroizi=[]
a=0
b=0

cent_index=random.sample(range(len(point_array)), 3)

for i in range(3):
    centroizi.extend([point_array[cent_index[i]]])


while(True):
  
  
    for i in range(len(point_array)):
        for j in range(len(centroizi)):
               
            x1=centroizi[j][0]
            y1=centroizi[j][1]
            x2=point_array[i][0]
            y2=point_array[i][1]

            d=abs(x1-x2)+abs(y1-y2)
            d_array.append(d)
            d_min=np.min(d_array)
            index=d_array.index(d_min)
    
        cluster[index].append(point_array[i])
        d_array.clear()
  
    
    print("\nCentroizi:")
    for k in range(len(centroizi)):
        for l in range(len(cluster[k])):
      
            a+=cluster[k][l][0]
            b+=cluster[k][l][1]
     
        x=a/len(cluster[k])
        y=b/len(cluster[k])
        centroizi[k]=[x,y]
       

        a=0
        b=0
    
    print(centroizi)
    
    for cl in range(len(cluster)):
        print("\nCluster-ul c"+str(cl+1)+ " este format din:",cluster[cl])
       
        
        cluster[cl].clear()
        print()
    if centroizi==centroizi_aux:
        break
    else:
        centroizi_aux.clear()
        centroizi_aux.extend(centroizi)


```
