```python
#Minimizarea unui sir binar folosind algoritmi genetici 

import numpy as np
import math
import random
f=[]
pi=[]
siruri=[]
array_z=[]
array_b=[]



#Initializare
def Initializare():
    for i in range(6):
        s=np.random.choice([0, 1], size=(6))
        val_b=s[0]*np.power(10,5)+s[1]*np.power(10,4)+s[2]*np.power(10,3)+s[3]*np.power(10,2)+s[4]*np.power(10,1)+s[5]*np.power(10,0)
        val_z=s[0]*np.power(2,5)+s[1]*np.power(2,4)+s[2]*np.power(2,3)+s[3]*np.power(2,2)+s[4]*np.power(2,1)+s[5]*np.power(2,0)
        if val_z==0:
            f.append(1)
        else:
            f.append(math.pow(val_z,-1))
       

        if len(str(val_b))!=6:
            val_b="0"*(6-len(str(val_b)))+str(val_b)
        else:
            val_b=str(val_b)
     
        array_b.append(val_b)
        array_z.append(val_z)
        siruri.append(s)
    for i in range(6):   
        pi.append(f[i]/sum(f))

    print(siruri)
    print("Siruri binare:",array_b)
    print("Valorile zecimale ale sirului:",array_z)
    print("Functiile obiectiv:",f)
    print("Functiile obiectiv relative:",pi)




#Selectie

def Selectie(populatie):
    f.clear()
    pi.clear()
    for s in populatie:
        
        val=int(s, 2)   
        if val==0:
            f.append(1)
        else:
            f.append(math.pow(val,-1))
    for i in range(6):   
        pi.append(f[i]/sum(f))

    populatie=np.random.choice(populatie, size=(6),p=pi)
    # print("\nPopulatia dupa selectie",array_b)


def incrucisare(populatie):
    lista_bin=list(populatie)
    for i in range(int(len(populatie)/2)):
        index_t=random.randint(1,6)
        perechi=np.sort(random.sample(range(len(lista_bin)),2))[::-1]
      
        sir1=populatie[perechi[0]]
        sir2=populatie[perechi[1]]
 
        
        aux=sir1
        sir1=sir1[0:index_t]+sir2[index_t:len(sir2)]
        sir2=sir2[0:index_t]+aux[index_t:len(aux)]

        populatie[perechi[0]]=sir1
        populatie[perechi[1]]=sir2
        

        lista_bin.pop(perechi[0])
        lista_bin.pop(perechi[1])

        
    # print("Populatia dupa incrucisare:",populatie)

#Mutatie
def mutatie(populatie):
    nr_mutatii=0
    for i in range (len(populatie)):
        for j in range (len(populatie[i])):
            rand=random.randrange(1,1000)
            # print(a)
            
            if rand==1:
                nr_mutatii+=1
                bit=populatie[i][j]
                # print ("Mutatie:")
                if bit=='0':
                    bit='1'
                elif bit=='1':
                    bit='0'
                # print (bit)
                populatie[i]= str(populatie[i][0:(j)])+str(bit)+str(populatie[i][j+1:len(populatie[i])])
        



Initializare()
while True:
    Selectie(array_b)
    incrucisare(array_b)
    mutatie(array_b)
    if sum(f)>=3:
        print("\nSuma functiilor obiectiv:",sum(f))
        print(array_b)
        break
 
```
